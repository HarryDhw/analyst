```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def gamedie(num):
    die = range(1,7)
    one = []
    two = []
    three = []
    four = []
    five = []
    six = []
    small = []
    big = []
    spray = []
    for i in range(num):
        a = np.random.choice(die)
        b = np.random.choice(die)
        c = np.random.choice(die)
        d = a+b+c
        spray.append(d)
        if a==b==c==1:
            one.append(d)
        elif a==b==c==2:
            two.append(d)
        elif a==b==c==3:
            three.append(d)
        elif a==b==c==4:
            four.append(d)
        elif a==b==c==5:
            five.append(d)
        elif a==b==c==6:
            six.append(d)
        elif d <= 10:
            small.append(d)
        else:
            big.append(d)
    return one,two,three,four,five,six,small,big,spray
```

```python
one,two,three,four,five,six,small,big,spray = gamedie(1000000)
```
